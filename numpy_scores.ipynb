```python
import numpy as np

# Task 1
np.random.seed(42)
scores = np.random.randint(50, 101, size=(5, 4))
print("Scores:\n", scores)

print("3rd student, 2nd subject:", scores[2, 1])
print("Last 2 students:\n", scores[3:, :])
print("First 3 students, subjects 2 and 3:\n", scores[:3, 1:3])
```

```python
col_mean = np.round(np.mean(scores, axis=0), 2)
print("Column-wise mean:", col_mean)

curve = np.array([5, 3, 7, 2])
curved_scores = scores + curve
curved_scores = np.clip(curved_scores, None, 100)
print("Curved scores:\n", curved_scores)

row_max = np.max(curved_scores, axis=1)
print("Row-wise max:", row_max)
```

```python
row_min = curved_scores.min(axis=1, keepdims=True)
row_max2 = curved_scores.max(axis=1, keepdims=True)
normalized = (curved_scores - row_min) / (row_max2 - row_min)
print("Normalized:\n", normalized)

idx = np.unravel_index(np.argmax(normalized), normalized.shape)
print("Student index:", idx[0], "Subject index:", idx[1])

above_90 = curved_scores[curved_scores > 90]
print("Scores strictly above 90:", above_90)
```
